Soal 1: Visualisasikan distribusi (KDE plot) dari kolom ```bill_length_mm``` dan kolom ```bill_depth_mm``` yang dikelompokkan berdasarkan kolom ```island```. Visualisasi dari ```bill_length_mm``` berada di kiri, sedangkan visualisasi ```bill_depth_mm``` berada di sebelah kanan (sub plots). Buatlah plot se aestetik mungkin. 

Berilah judul pada masing-masing sub-plot, kemudian berilah judul utama di tengah.

````python
df = sns.load_dataset("penguins")
````

````python
df = df.dropna().reset_index(drop = True)
````

````python
df.isna().sum()
````

````python
df
````

````python
plot1 = sns.kdeplot(data=df, x="bill_length_mm", hue="island", fill=True, common_norm=False, 
                    palette="pastel", alpha=.5, linewidth=0)
````

````python
plot2 = sns.kdeplot(data=df, x="bill_depth_mm", hue="island", fill=True, common_norm=False, 
                    palette="pastel", alpha=.5, linewidth=0)
````

````python
fig, ax = plt.subplots(1,2,figsize=(15,5))

plot1 = sns.kdeplot(data=df, x="bill_length_mm", hue="island", fill=True, common_norm=False, 
                    palette="pastel", alpha=.5, linewidth=0,ax=ax[0])

plot2 = sns.kdeplot(data=df, x="bill_depth_mm", hue="island", fill=True, common_norm=False, 
                    palette="pastel", alpha=.5, linewidth=0,ax=ax[1])

fig.suptitle('Size variation of Penguins Bill on each Island', fontsize=16)
plot1.title.set_text('Bill length (mm)')
plot2.title.set_text('Bill depth (mm)')

plt.tight_layout(pad=5.0)
````

Soal 2: Visualisasikan scatterplot dengan ketentuan sebagai berikut:
- bentuk : subplot dengan 1 column dan 2 row.

    - sumbu y : flipper_length_mm
- subplot 2:

````python
fig, ax = plt.subplots(2,1,figsize=(10,8))

sns.scatterplot(data=df, x='body_mass_g', y='flipper_length_mm', 
                hue='island',palette="pastel", s=100, ax=ax[0])

sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', 
                hue='island', palette="pastel",s=100, ax=ax[1])
````

Interpretasi:

Berdasarkan scatter plot yang dihasilkan, banyak spesies penguin yang berasal dari pulau Biscoe memiliki berat badan yang beratnya melebihi berat spesies penguin dari pulau lainnya. Begitu pula dengan panjang sirip (flipper) dan panjang paruhnya juga melebihi panjang spesies penguin dari pulau lainnya. Sedangkan spesies penguin dari pulau Torgersen dan Dream memiliki postur tubuh (sirip dan paruh) serta berat badan yang rata-rata dimiliki semua spesies dari seluruh pulau (Torgersen, Biscoe, dan Dream).
